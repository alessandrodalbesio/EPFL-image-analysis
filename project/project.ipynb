Tiling puzzles are a classic type of puzzle game that consists of fitting together pieces of a given shape (in this case squared to form a complete image. The goal of this project is to develop an algorithm that can automatically reconstruct tiling puzzles from a single input image.

````python
import cv2

# Plot the results for the presentation
def compare_processing_operations_image(original_image: np.ndarray, canny_image: np.ndarray, flood_fill_image: np.ndarray, original_image_with_contourn: np.ndarray):
    # Plot the comparison between the various phases of the segmentation process
    fig, ax = plt.subplots(1, 4, figsize=(20, 5))
    ax[0].imshow(original_image)
    ax[0].set_title("Original Image")
    ax[1].imshow(canny_image)
    ax[1].set_title("Original Image after Canny")
    ax[2].imshow(flood_fill_image)
    ax[2].set_title("Result of the flood fill")
    ax[3].imshow(original_image_with_contourn)
    ax[3].set_title("Bounding boxes")
    plt.show()

def compare_processing_operations_single_pieces():
    pass

def bc_image(image):
    # Copy the image
    img_copy = image.copy()

    # Perform a flood fill starting from the corners
    h, w = image.shape[:2]
    mask_fill = np.zeros((h + 2, w + 2), np.uint8)

    # Define the starting points criteria
    starting_points = [(0, 0), (0, 1999), (1999, 0), (1999, 1999)]
    for point in starting_points:
        cv2.floodFill(img_copy, mask_fill, point, 0)

    # Return the background
    return mask_fill

def rotate_image(image, angle_rotation):
    # Rotate the image
    rows, cols = image.shape[:2]
    M = cv2.getRotationMatrix2D((cols/2, rows/2), angle_rotation, 1)
    image = cv2.warpAffine(image, M, (cols, rows))
    return image

NUMBER_OF_PIECES = {
                    '0': 28,
                    '1': 21, 
                    '2': 28, 
                    '3': 21,
                    '4': 20,
                    '5': 28,
                    '6': 29,
                    '7': 28,
                    '8': 27,
                    '9': 29,
                    '10': 28,
                    '11': 19
                    }

# Segmentation with contours 
def segmentation(image: np.ndarray, plot_for_presentation: bool = False) -> list[np.ndarray]:
    ### Identification of the pieces of the puzzle in the big image ###
    
    # Apply Canny to the image
    img_canny = cv2.Canny(image, 50, 200)

    # Apply dilatation to close the edges
    kernel = np.ones((5, 5), np.uint8)
    img_canny = cv2.dilate(img_canny, kernel, iterations=1)

    # Get the background isolated
    bc = bc_image(img_canny)

    # Find contours on the bc image
    contours, hierarchy = cv2.findContours(bc, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)            
    contours = [c for c, h in zip(contours, hierarchy[0]) if h[3] != -1]

    # Draw the contours on the original image
    image_with_contours = image.copy()
    cv2.drawContours(image_with_contours, contours, -1, (255, 0, 0), 5)

    if plot_for_presentation:
        compare_processing_operations_image(image, img_canny, bc, image_with_contours)


    ### Extraction of the pieces of the puzzle, rotation and cropping ###
    # This part is not working correctly -> cuts out too many images
    
    images = []
    images_to_plot = []
    for contour in contours:
        # Get the bounding box
        x, y, w, h = cv2.boundingRect(contour)

        # Crop the image
        imageOfInterest = image[y:y+h, x:x+w]

        # Compute the angle of rotation
        angle = cv2.minAreaRect(contour)[2]

        # Rotate the image
        imageOfInterest = rotate_image(imageOfInterest, angle)
        
        imageOfInterestBlackAndWhite = cv2.cvtColor(imageOfInterest, cv2.COLOR_BGR2GRAY)
        imageOfInterestBlackAndWhite[imageOfInterestBlackAndWhite > 0] = 255
        kernel = np.ones((9,9), np.uint8)
        imageOfInterestBlackAndWhite = cv2.erode(imageOfInterestBlackAndWhite, kernel, iterations=1)

        # Apply Canny   
        imageOfInterestCanny = cv2.Canny(imageOfInterest, 0, 200)

        # Keep a pixel only if it is white in both images
        imageOfInterestCanny = imageOfInterestBlackAndWhite & imageOfInterestCanny

        # Crop the first and last row and column
        imageOfInterestCanny = imageOfInterestCanny[1:-1, 1:-1] == 255


        # Crop the images such that only the piece of the puzzle is left
        THRESH = 10
        cropValueLeft = 0
        cropValueRight = 0
        cropValueTop = 0
        cropValueBottom = 0

        for i in range(imageOfInterestCanny.shape[1]):
            if np.sum(imageOfInterestCanny[:, i]) > THRESH:
                cropValueLeft = i + 1
                break
        for i in range(imageOfInterestCanny.shape[1] - 1, -1, -1):
            if np.sum(imageOfInterestCanny[:, i]) > THRESH:
                cropValueRight = i
                break
        for i in range(imageOfInterestCanny.shape[0]):
            if np.sum(imageOfInterestCanny[i, :]) > THRESH:
                cropValueTop = i + 1
                break
        for i in range(imageOfInterestCanny.shape[0] - 1, -1, -1):
            if np.sum(imageOfInterestCanny[i, :]) > THRESH:
                cropValueBottom = i
                break
        

        # Crop the image        
        imageOfInterest = imageOfInterest[cropValueTop:cropValueBottom, cropValueLeft:cropValueRight]

        # Verify that the size of the image is not zero
        if imageOfInterest.shape[0] >= 10 and imageOfInterest.shape[1] >= 10:
            images.append(imageOfInterest)

    return images
````

````python
def save_solution_puzzles(image_index: int, solved_puzzles: list[np.array], outliers: list[np.array], folder: str = DEFAULT_SAVE_FOLDER) -> None:
    """Description:
    ----------
    Function that saves the various puzzles and outliers in the folder data_project/train_solution_XX (where XX is the group number) in the following format:
        - solution_XX_YY.png (where XX is the image index and YY is the puzzle index)
        - outlier_XX_YY.png (where XX is the image index and YY is the outlier index)

    Args:
    ----------
        - image_index (int): index of the image to save (train_XX.png)
        - solved_puzzles (list[np.array]): list containing the saved puzzles (each element of the list rappresents a solved puzzle)
        - outliers (list[np.array]): list containing the outliers (each outlier is a numpy array 128 x 128 x 3)
        - folder (str, optional): Initial name of the folder in which the solutions will be saved (folder'_solution_5') . Default to "data_project/train" (solutions will be saved in data_project/train_solution_5)
    """
    path_solution = os.path.join(folder + "_solution_{}".format(str(GROUP_ID).zfill(2)))
    if not os.path.isdir(path_solution):
        os.mkdir(path_solution)

    for i, puzzle in enumerate(solved_puzzles):
        filename = os.path.join(path_solution, "solution_{}_{}.png".format(str(image_index).zfill(2), str(i).zfill(2)))
        Image.fromarray(puzzle).save(filename)

    for i , outlier in enumerate(outliers):
        filename =os.path.join(path_solution, "outlier_{}_{}.png".format(str(image_index).zfill(2), str(i).zfill(2)))
        Image.fromarray(outlier).save(filename)
````
